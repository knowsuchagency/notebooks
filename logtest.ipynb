```python
from pathlib import Path
import logging

logfile = Path('./logtest.log')
if logfile.exists():
    logfile.unlink()

# create logger
logger = logging.getLogger('logtest.log')
logger.setLevel(logging.DEBUG)

# create file handler and add it
fh = logging.FileHandler('logtest.log')
fh.setLevel(logging.DEBUG)
logger.addHandler(fh)

# create and add formatter
formatter = logging.Formatter('%(asctime)s:%(name)s:%(lineno)s:%(msg)s')
fh.setFormatter(formatter)

# print
logger.debug('foo')

with open('logtest.log') as logfile:
    print(logfile.read())
```
